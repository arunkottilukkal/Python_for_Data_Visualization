```python
%matplotlib inline

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('files/Election_2016.xlsx')
newDF = df.groupby('Party', axis=0).sum()
newDF['Vote'].plot(kind='pie')
plt.title('TN ELECTION RESULTS 2014')
plt.legend(frameon=False, loc='lower center', ncol=5)
plt.show()
```
