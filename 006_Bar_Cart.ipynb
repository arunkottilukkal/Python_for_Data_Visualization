```python
%matplotlib inline
# %matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('files/World_Migration_2019.xlsx')
newDF = df[['Year','India','Pakistan','Destination']]
newDF = newDF[(newDF['India']>0) & (newDF['Destination']=="United States of America")]

xAxis = newDF['Year'].astype(str)
yAxis = newDF['India'].astype(int)
y1Axis = newDF['Pakistan'].astype(int)

plt.style.use('ggplot')
plt.bar(xAxis, yAxis, color='blue',label="India")
plt.bar(xAxis, y1Axis, color='green',label="Pakistan")
plt.title('US Immigration 1990 - 2019')
plt.xlabel('Year')
plt.ylabel('No(Immigrants)')
plt.legend()
plt.show()
```
