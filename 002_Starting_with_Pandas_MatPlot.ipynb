```python
%matplotlib inline
%matplotlib notebook

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('files/UN_world_crime_record_2019.csv')
country_name = ["India","Pakistan"]
df_ind_vs_pak = df[df.CountryName.isin(country_name)]
years = list(map(str,[2005,2010,2013,2016]))
ind_pop = [1144100000,1231000000,1278600000,1324200000]
pak_pop = [153900000,170600000,181700000,193200000]
print(years)
# df_ind_vs_pak.loc[['India',years]].plot(kind='line')
# plt.show()             
# df_ind_vs_pak.head()
series = set(df_ind_vs_pak['Series'])

ind_crime = []
pak_crime = []

# for val in series:
#     for y in years:
#         ifilter = df_ind_vs_pak[(df_ind_vs_pak['CountryName']=="India") & (df['Series']==val) &(df['Year']==int(y))]
#         total = ifilter['Value'].sum()
#         print(y,val, total)

# 114.41
# 123.1
# 127.86

for i, y in enumerate(years):
    ifilter = df_ind_vs_pak[(df_ind_vs_pak['CountryName']=="India") & (df['Year']==int(y))]
    popVal = ind_pop[i]/100000
    totalCrime = (ifilter['Value'].sum()/popVal)*100
    ind_crime.append(totalCrime)
    print(y, totalCrime)
    
for i, y in enumerate(years):
    ifilter = df_ind_vs_pak[(df_ind_vs_pak['CountryName']=="Pakistan") & (df['Year']==int(y))]
    popVal = pak_pop[i]/100000
    totalCrime = (ifilter['Value'].sum()/popVal)*100
    pak_crime.append(totalCrime)
    print(y, totalCrime)
    
plt.plot(years, pak_crime, color='g')
plt.plot(years, ind_crime, color='blue')
plt.xlabel('Countries')
plt.ylabel('Crime rate')
plt.title('India vs Pakistan Crime Rate')
plt.show()
```
