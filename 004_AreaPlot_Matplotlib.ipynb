```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('files/World_Migration_2019.xlsx')

dfNew = df[df['India']!=""]
newDF = dfNew[['Year','Destination','India','Pakistan']]
newDF['India'].fillna(0, inplace=True)
newDF['Pakistan'].fillna(0, inplace=True)
# newDF.sort_values(['India','Pakistan'],ascending=False,axis=0,inplace=True)
newDF.sort_values(["India", "Pakistan"], axis=0, ascending=False, inplace=True) 
DFrame = newDF.head(10)
# years = list(map(str,range(1990,2020)))
# DFrame.head()
# DFrame = DFrame.transpose()
DFrame.head()
DFrame.plot(kind='area')
plt.title('IND vs PAK 1990 - 2019 TOP 10 IMMIGRATIONS')
plt.ylabel("No(Immigrants)")
plt.xlabel("Years")
plt.show()
```
