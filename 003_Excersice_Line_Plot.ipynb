```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('files/World_Migration_2019.xlsx')
filterCountry = ['United States of America']
US_IMMEGRANTS = df[df['Destination']=='United States of America']
YEAR = US_IMMEGRANTS['Year']
INDIANS = US_IMMEGRANTS['India']
PAKISTANIES = US_IMMEGRANTS['Pakistan']
```

```python
plt.plot(YEAR,INDIANS,'b',label="Indians")
plt.plot(YEAR,PAKISTANIES,'g',label="Pakistanies")
plt.xlabel('Year')
plt.ylabel('Immigrants')
plt.title('India vs Pakistan > US Immigrants > 1990 - 2019')
plt.legend()
plt.show()
```
