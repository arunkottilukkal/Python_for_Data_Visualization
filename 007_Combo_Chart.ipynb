```python
%matplotlib inline

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('files/World_Migration_2019.xlsx')
newDF = df[['Year','Destination','India','Pakistan']]
newDF = newDF[newDF['Destination']=='Russian Federation']
newDataSet = newDF
```

```python
plt.bar(newDataSet['Year'].astype(str),newDataSet['Pakistan'],width=0.9,color='red',label='Pakistan')
x = newDataSet['Year'].astype(str)
y = newDataSet['India'].astype(int)
plt.plot(x,y,label="India")
plt.legend()
plt.title('India vs Pakistan Russian Immigrants')
plt.xlabel('Year')
plt.ylabel('n(Immigrants)')
plt.show()
```
